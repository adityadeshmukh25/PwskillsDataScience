# Flight Price Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
flight_data = pd.read_excel('flight_price.xlsx')  # Replace 'flight_price_dataset.csv' with your dataset file name

# Display dimensions of the dataset
print("Dimensions of the dataset (rows, columns):", flight_data.shape)
```

```python
# Create a histogram to visualize the distribution of flight prices
plt.figure(figsize=(8, 6))
plt.hist(flight_data['Price'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

```python
# Calculate the range, minimum, and maximum prices
price_range = flight_data['Price'].max() - flight_data['Price'].min()
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

print("Price Range:", price_range)
print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
```

```python
# Create a boxplot to compare prices by airline
plt.figure(figsize=(10, 6))
plt.boxplot([flight_data[flight_data['Airline'] == airline]['Price'] for airline in flight_data['Airline'].unique()],
            labels=flight_data['Airline'].unique())
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Price Variation by Airline')
plt.xticks(rotation=90)
plt.show()
```

```python
# Identify potential outliers using a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(flight_data['Price'], vert=False)
plt.xlabel('Price')
plt.title('Identification of Potential Outliers')
plt.show()
```

# Google Playstore dataset

```python
import pandas as pd

# Load the dataset
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')  # Replace 'google_playstore_dataset.csv' with your dataset file name

# Display dimensions of the dataset
print("Dimensions of the Google Playstore dataset (rows, columns):", playstore_data.shape)

# Display dimensions of the dataset
print("Dimensions of the Google Playstore dataset (rows, columns):", playstore_data.shape)
```

```python
import matplotlib.pyplot as plt

# Create a boxplot to compare app ratings by category
plt.figure(figsize=(10, 6))
playstore_data.boxplot(column='Rating', by='Category', figsize=(14,8), rot=90)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('App Ratings Variation by Category')
plt.show()
```

```python
# Check for missing values
missing_values = playstore_data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(8, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('App Size vs Rating')
plt.show()
```

```python
# Create a bar chart to compare average prices by app type
avg_prices = playstore_data.groupby('Type')['Price'].mean()
plt.figure(figsize=(8, 6))
avg_prices.plot(kind='bar', color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
plt.show()
```
