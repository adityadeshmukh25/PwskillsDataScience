A1: Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy⁴. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK⁴. It is used for data visualization and creating static, animated, and interactive visualizations in Python¹. Some of the plots that can be plotted using the Pyplot module of Matplotlib are:

- Line plot: A plot that shows the relationship between two variables by drawing a line connecting the data points.
- Bar plot: A plot that shows the frequency or count of different categories using bars of different heights.
- Histogram: A plot that shows the distribution of a numerical variable by grouping the data into bins and counting the number of observations in each bin.
- Pie chart: A plot that shows the proportion of different categories in a whole by dividing a circle into sectors.
- Box plot: A plot that shows the summary statistics of a numerical variable by drawing a box with whiskers that indicate the minimum, maximum, median, and quartiles of the data.



A2: A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data⁷. It is used to observe and show relationships between two numeric variables⁶. If the points are coded (color/shape/size), one additional variable can be displayed⁷.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Random seed Graph')
plt.scatter(x,y)
plt.show()
```

A3: The subplot() function is used to create multiple plots in the same figure. It takes three arguments that describe the layout of the figure: the number of rows, the number of columns, and the index of the current plot. The index starts from 1 in the upper left corner and increases to the right. For example, subplot(2, 2, 1) means that the figure has 2 rows and 2 columns, and the current plot is the first one in the upper left corner¹.

To draw four line plots using the subplot() function, we can use the following code:




```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
plt.figure(figsize=(10,10))

# Plot line 1 in the first subplot
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Line Plot 1")
plt.xlabel("x")
plt.ylabel("y")

# Plot line 2 in the second subplot
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Line Plot 2")
plt.xlabel("x")
plt.ylabel("y")

# Plot line 3 in the third subplot
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Line Plot 3")
plt.xlabel("x")
plt.ylabel("y")

# Plot line 4 in the fourth subplot
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Line Plot 4")
plt.xlabel("x")
plt.ylabel("y")

# Show the figure
plt.show()
```

A4: A bar plot is a type of plot that shows the frequency or count of different categories using bars of different heights⁶. It is used to compare the values of different categories or groups⁷. For example, we can use a bar plot to compare the profit of different companies.

To plot a bar plot and a horizontal bar plot using the given data, we can use the following code:



```python
import matplotlib.pyplot as plt
import numpy as np

# Data for bar plots
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure with two subplots
plt.figure(figsize=(10,10))

# Plot a vertical bar plot in the first subplot
plt.subplot(2,1,1)
plt.bar(company,profit)
plt.title("Vertical Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")

# Plot a horizontal bar plot in the second subplot
plt.subplot(2,1,2)
plt.barh(company,profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")

# Show the figure
plt.show()
```


A5: A box plot is a type of plot that shows the summary statistics of a numerical variable by drawing a box with whiskers that indicate the minimum, maximum, median and quartiles of the data⁹. It is used to observe and show the distribution and variability of a numerical variable[^10^]. For example,
we can use a box plot to compare the distribution of two sets of data.

To plot a box plot using the given data, we can use the following code:




```python
import matplotlib.pyplot as plt
import numpy as np

# Data for box plot
box1 = np.random.normal(100,10 ,200)
box2 = np.random.normal(90 ,20 ,200)

# Create a figure with one subplot
plt.figure(figsize=(10 ,10))

# Plot a box plot with two boxes
plt.boxplot([box1 ,box2])
plt.title("Box Plot")
plt.xlabel("Box Number")
plt.ylabel("Value")

# Show the figure
plt.show()
```
