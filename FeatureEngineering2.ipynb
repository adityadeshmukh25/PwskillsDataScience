### Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

Min-Max scaling is a technique used in data preprocessing to transform features by scaling them to a specific range, usually between 0 and 1. The formula for Min-Max scaling is:

\[ X_{\text{scaled}} = \frac{X - \text{min}(X)}{\text{max}(X) - \text{min}(X)} \]

This method is useful when the features have different units or scales. For example, if you have a dataset with values ranging from 10 to 100, Min-Max scaling will transform these values to the range [0, 1].

Example:
Consider a dataset with a feature 'Income' that ranges from \$20,000 to \$100,000. Applying Min-Max scaling to this feature will transform the values to a range between 0 and 1, making the minimum value 0 and the maximum value 1.

### Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling? Provide an example to illustrate its application.

The Unit Vector technique, also known as normalization, scales individual samples to have a Euclidean length of 1. This means that it scales the values within each sample (row) rather than across features. It's different from Min-Max scaling as it normalizes the rows instead of columns and maintains the direction of the data points.

Example:
Consider a dataset where rows represent vectors. Applying Unit Vector scaling will normalize each row vector to have a length of 1.

### Q3. What is PCA (Principal Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

PCA is a technique used for dimensionality reduction in which it identifies the most important features (principal components) in a dataset while retaining most of the variance. It achieves this by transforming the original features into a new set of uncorrelated features called principal components.

Example:
In a dataset containing multiple correlated features like height, weight, age, and income, PCA can identify the most significant patterns and represent the data using fewer principal components that capture most of the variance.

### Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature Extraction? Provide an example to illustrate this concept.

PCA is a method used for feature extraction by transforming the original features into a set of new features (principal components). It helps in reducing the dimensionality of the dataset while retaining most of the relevant information.

Example:
In a dataset with multiple correlated features, applying PCA will create new orthogonal features (principal components) that capture the maximum variance in the data. These principal components can then be used as the reduced set of features for modeling.

   \[ X_{\text{scaled}} = \frac{X - \text{min}(X)}{\text{max}(X) - \text{min}(X)} \times ( \text{new max} - \text{new min}) + \text{new min} \]
